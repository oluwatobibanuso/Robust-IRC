```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# read images, videos and webcam 
img = cv2.imread('lena.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
cv2.imshow("Output",img)
cv2.waitKey(1000) 
```

```python
cap = cv2.VideoCapture("test_vid.mp4")

while True:
    success, img_vid = cap.read()
    cv2.imshow("Video",img_vid)
    if cv2.waitKey(1000) & 0xFF == ord('q'):
       break
```

```python
# basic functionalities 
img = cv2.imread('lena.png')
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Lena Gray", imgGray)
cv2.waitKey(10000)
```

```python
imgBlur = cv2.GaussianBlur(imgGray,(7,7),0)
cv2.imshow('Lena Blur', imgBlur)
cv2.waitKey(1000)
```

```python
# edge detecting
imgCanny = cv2.Canny(img,150,200)
cv2.imshow('Lena Edges',imgCanny)
cv2.waitKey(100)
```

```python
#image dilation
kernel = np.ones((5,5),np.uint8)
imgDilation = cv2.dilate(imgCanny,kernel,iterations=1)
cv2.imshow("Lena Dilation",imgDilation)
cv2.waitKey(1000)
```

```python
#image erosion
imgEroded = cv2.erode(imgDilation,kernel,iterations=1)
cv2.imshow("Lena Eroded",imgEroded)
cv2.waitKey(1000)
```

```python
#resize images 
img = cv2.imread('lambo.png')
print(img.shape)
print("Height : {} pixels".format(img.shape[0]))
print("Weight : {} pixels".format(img.shape[1]))
print("Channels : {}".format(img.shape[2]))
```

```python
#resize image
imgResize = cv2.resize(img,(300,200))
cv2.imshow("Lambo Resized",imgResize)
cv2.waitKey(1000)
```

```python
#cropping images 
imgCrop = img[0:200,200:500]
cv2.imshow("Cropped",imgCrop)
cv2.waitKey(1000)
```

```python
img_np = np.zeros((512,512,3),np.uint8)
img_np[:] = 255,0,0
cv2.line(img_np,(0,0),(500,500),(0,255,0),3)
cv2.rectangle(img_np,(0,0),(400,300),(0,255,0),3)
cv2.imshow("Numpy Image",img_np)
cv2.waitKey(1000)
```
